Write a function: input a number x and return x^2+3x+1

```r
quad=function(x)
{
  return(x^2+3*x+1)
}
quad(2)
```

Write a function: input a vector and return the product of all the entries in the vector

```r
product=function(x){
  prod=1
  for(i in 1:length(x)){
    prod=prod*x[i]
  }
  return(prod)
}
product(c(1,3,5))
```

Write a function: input a vector v and a number n and return a vector with its missing data being replaced by n

```r
missvect=function(x,n)
{
  x[is.na(x)]=n
  return(x)
}
missvect(c(1,NA,NA,9),5)
```

Write a function: input a numeric vector and return a vector with its missing data being replaced by the mean of the vector

```r
miss_mean=function(x)
{
  x[is.na(x)]=mean(x, na.rm=TRUE)
  return(x)
}
miss_mean(c(1,3,NA,8))
```

Write a function: input a data frame and return a data frame with numeric missing values being replaced by the mean of the corresponding column.  Check your function on the titanic data

```r
miss_data=function(x){
  for (i in 1:ncol(x)){
    if (is.numeric(x[,i])){
      x[,i][is.na(x[,i])]=mean(x[,i], na.rm=TRUE)
    }
  }
  return (x)
}
miss_data(titanic)
```

Write a function that plots a bar chart of an input categorical variable.  Check your function on the titanic data

```r
bar_chart=function(df,var)
{
  library(ggplot2)
  ggplot(data=df)+geom_bar(mapping=aes(x=var))
}
bar_chart(titanic,titanic$Survived)
```

Write a function that plots bar chart of two input categorical variables.  Check your function on the titanic data

```r
bar_chart2=function(block,var1,var2)
{
  library(ggplot2)
  ggplot(data=block)+geom_bar(mapping=aes(x=var1, fill=var2), position="dodge")
}

bar_chart2(titanic,titanic$Survived,titanic$Sex)
```
