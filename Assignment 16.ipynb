Use simulation to verify that the sum squares of k standard normal distributions is a Chi-square distribution with degree of freedom k. 

```r
n=1000

x=rnorm(n,0,1)
y=rnorm(n,0,1)
z=rnorm(n,0,1)
w=rnorm(n,0,1)

ss=x^2+y^2+z^2+w^2

hist(ss)
```

A fair dice is toss three times.  Use simulation to find the probability that the number in the first toss is greater than the sum of the number in the second toss and the third toss. 

```r
n=1000
s1=sample(1:6,n,replace = TRUE)
s2=sample(1:6,n,replace = TRUE)
s3=sample(1:6,n,replace = TRUE)
k=sum(s1>s2+s3)
k/n
```

Let U1,U2,U3 all come from a uniform(0,1) distribution. Let M=max(U1,U2,U3). Estimate the probability P(M>0.75).

```r
n=1000
M=c()
for (i in 1:n){
  u1=runif(1,0,1)
  u2=runif(1,0,1)
  u3=runif(1,0,1)
  M=c(M,max(u1,u2,u3))
}

k=sum(M>0.75)
k/n
```

Generate 1000 pair of samples (x, y) from random variables (X, Y) such that. 
-	X follows uniform distribution (-1,1)
-	Y = 2X+3+e
-	e is an error following normal distribution with the mean 0 and standard deviation s = 1. 
Find the slope and intercept of the fitted line and compare them with the true values.

```r
n=1000
x=runif(n,-1,1)
e=rnorm(n,0,1)
y=2*x+3+e
line=lm(y~x)
line
plot(line)
```
