Reading in the Titanic training data set, correcting certain variable types, and loading in the ggplot2 library for visualizations

```r
titanic=read.csv("train.csv")
titanic$Survived=factor(titanic$Survived)
titanic$Pclass=factor(titanic$Pclass)
titanic$PassengerId=factor(titanic$PassengerId)
library(ggplot2)
```

This scatter plot of age and fare shows that there isn't a relationship between the 2 variables, and the people that paid the highest fare were in their mid-30s.

```r
ggplot(data=titanic)+geom_point(mapping=aes(x=Age, y=Fare))
```

This bar chart of Pclass shows that the majority of passengers on the Titanic were in 3rd class.

```r
ggplot(data=titanic)+geom_bar(mapping=aes(x=Pclass))
```

This boxplot of gender and age shows that males were slightly older than females, and also had more outlier ages

```r
ggplot(data=titanic)+geom_boxplot(mapping=aes(x=Sex, y=Age))
```

This bar chart shows that more females survived than males did.

```r
ggplot(data=titanic)+geom_bar(mapping=aes(x=Sex, fill=Survived),position="dodge")
```

This density chart shows that the majority of people who died were between the ages of 20 and 30 years old.

```r
ggplot(data=titanic)+geom_density(mapping=aes(x=Age, color=Survived))
```

This histogram shows that the majority of the passengers on the Titanic were in their early-20s.

```r
ggplot(data=titanic)+geom_histogram(mapping=aes(x=Age))
```

This bar chart shows that the majority of passengers in 3rd class died, while a majority of passengers in 1st class survived

```r
ggplot(data=titanic)+geom_bar(mapping=aes(x=Pclass, fill=Survived))
```
